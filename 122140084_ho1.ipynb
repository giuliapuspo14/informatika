## 1. Pendahuluan

Pada tugas ini, kita akan membuat beberapa sinyal **sinus** dan **cosinus** dengan berbagai parameter (amplitudo, frekuensi, dan fase). Tujuan utama dari tugas ini adalah untuk memahami:

1. **Membuat Gelombang Sinus dan Cosinus**:
   - Buat array `t` yang berkisar dari 0 hingga 2 dengan langkah sebesar 0.0001.
   - Buat sinyal-sinyal berikut menggunakan array waktu ini:
     1. $y1 = 2 \cdot \sin(2 \pi \cdot 3 \cdot t + 0)$
     2. $y2 = 1 \cdot \cos(2 \pi \cdot 4 \cdot t + \pi/4)$
     3. $y3 = -1 \cdot \sin(2 \pi \cdot 5 \cdot t + \pi/2)$
     4. $y4 = 0.5 \cdot \cos(2 \pi \cdot 6 \cdot t + \pi)$

2. **Perbandingan Subplot**:
   - Buat gambar dengan 4 subplot (grid 2x2) untuk membandingkan semua sinyal secara berdampingan.
   - Setiap subplot harus berisi salah satu sinyal \( y1 \), \( y2 \), \( y3 \), dan \( y4 \).

3. **Pertanyaan Analisis**:
   - Jawab pertanyaan berikut dalam sel markdown di Jupyter Notebook Anda:
     1. Berapa amplitudo dan frekuensi masing-masing sinyal?
     2. Bagaimana pergeseran fase mempengaruhi posisi gelombang? (Anda dapat mengubah nilai fase pada sinyal-sinyal yang telah dibuat sesuka anda)
     3. Bandingkan sinyal-sinyal dengan amplitudo yang berbeda dan diskusikan bagaimana amplitudo mempengaruhi tampilan gelombang.
     4. Bandingkan sinyal-sinyal dengan pergeseran fase yang berbeda dan diskusikan bagaimana pergeseran fase mempengaruhi tampilan gelombang.

4. **Tugas Lanjutan**:
   - Buat sinyal baru $y5$ yang merupakan kombinasi dari $y1$ dan $y2$ yaitu, $y5 = y1 + y2$.
   - Plot $y5$ dan diskusikan bagaimana kombinasi dua gelombang sinus/cosinus mempengaruhi bentuk gelombang yang dihasilkan.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Membuat array waktu t dari 0 hingga 2 dengan langkah 0.0001
t = np.arange(0, 2, 0.0001)

# 2. Mendefinisikan sinyal-sinyal sesuai instruksi
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)
```

## Menampilkan Sinyal dalam Subplot (2x2)

Kita akan membuat 4 subplot untuk masing-masing sinyal. Hal ini memudahkan kita membandingkan keempat sinyal secara visual.


```python
#Membuat Subplot 2x2
fig, axs =plt.subplots(2, 2, figsize=(10, 6))


# Subplot untuk y1
axs[0, 0].plot(t, y1, color='blue')
axs[0, 0].set_title('Sinyal y1 = 2 sin(2π·3·t)')
axs[0, 0].set_xlabel('Waktu (s)')
axs[0, 0].set_ylabel('Amplitudo')

# Subplot untuk y2
axs[0, 1].plot(t, y2, color='red')
axs[0, 1].set_title('Sinyal y2 = cos(2π·4·t + π/4)')
axs[0, 1].set_xlabel('Waktu (s)')
axs[0, 1].set_ylabel('Amplitudo')

# Subplot untuk y3
axs[1, 0].plot(t, y3, color='green')
axs[1, 0].set_title('Sinyal y3 = - sin(2π·5·t + π/2)')
axs[1, 0].set_xlabel('Waktu (s)')
axs[1, 0].set_ylabel('Amplitudo')

# Subplot untuk y4
axs[1, 1].plot(t, y4, color='magenta')
axs[1, 1].set_title('Sinyal y4 = 0.5 cos(2π·6·t + π)')
axs[1, 1].set_xlabel('Waktu (s)')
axs[1, 1].set_ylabel('Amplitudo')

plt.tight_layout()
plt.show()
```

## 3. Pertanyaan Analisis

### 3.1. Amplitudo dan Frekuensi

1. **\( y_1 \)**  
   - **Amplitudo:** 2  
   - **Frekuensi:** 3 Hz  
     *Penjelasan:* Frekuensi ditentukan oleh faktor yang mengalikan variabel waktu \( t \) dalam fungsi sinus atau cosinus. Dalam hal ini, faktor tersebut adalah 3, sehingga frekuensinya adalah 3 Hz.

2. **\( y_2 \)**  
   - **Amplitudo:** 1  
   - **Frekuensi:** 4 Hz  
     *Penjelasan:* Faktor pengali \( t \) adalah 4, sehingga frekuensinya adalah 4 Hz.

3. **\( y_3 \)**  
   - **Amplitudo:** 1 (dengan tanda negatif yang menunjukkan pembalikan fase)  
   - **Frekuensi:** 5 Hz  
     *Penjelasan:* Faktor pengali \( t \) adalah 5, sehingga frekuensinya adalah 5 Hz.

4. **\( y_4 \)**  
   - **Amplitudo:** 0.5  
   - **Frekuensi:** 6 Hz  
     *Penjelasan:* Faktor pengali \( t \) adalah 6, sehingga frekuensinya adalah 6 Hz.

### 3.2. Pengaruh Pergeseran Fase

- **Pergeseran fase** mengacu pada pergeseran horizontal gelombang pada sumbu waktu. Pergeseran ini menentukan titik awal siklus gelombang.

- **Contoh:**  
  - Sinyal dengan **fase 0** akan memulai siklusnya pada titik standar (misalnya, \( \sin(0) = 0 \)).  
  - Sinyal dengan **fase \( \pi/4 \)** akan bergeser ke kiri atau ke kanan, sehingga puncak, lembah, dan titik nolnya tidak terjadi pada waktu yang sama dengan sinyal tanpa pergeseran fase.

- **Ilustrasi:**  
  - Misalkan terdapat dua sinyal sinus dengan frekuensi dan amplitudo yang sama, tetapi salah satunya memiliki pergeseran fase \( \pi/2 \). Sinyal dengan pergeseran fase akan mencapai puncaknya lebih awal atau lebih lambat dibandingkan sinyal tanpa pergeseran fase.

### 3.3. Pengaruh Amplitudo

- **Amplitudo** menentukan tinggi puncak dan kedalaman lembah gelombang. Amplitudo yang lebih besar menghasilkan puncak yang lebih tinggi dan lembah yang lebih dalam.

- **Contoh:**  
  - Sinyal \( y_1 \) dengan amplitudo 2 akan memiliki puncak pada nilai 2 dan lembah pada nilai -2.
  - Sinyal \( y_4 \) dengan amplitudo 0.5 akan memiliki puncak pada nilai 0.5 dan lembah pada nilai -0.5.

- **Pengaruh visual:**  
  - Sinyal dengan amplitudo lebih besar akan tampak lebih "tinggi" pada plot dibandingkan dengan sinyal beramplitudo lebih kecil.

### 3.4. Pengaruh Fase

- **Pergeseran fase** mempengaruhi posisi horizontal gelombang pada plot. Sinyal dengan pergeseran fase akan memiliki titik nol, puncak, dan lembah yang terjadi pada waktu yang berbeda dibandingkan dengan sinyal tanpa pergeseran fase.

- **Contoh:**  
  - Sinyal \( y_2 \) dengan pergeseran fase \( \pi/4 \) akan mencapai puncaknya lebih awal atau lebih lambat dibandingkan sinyal tanpa pergeseran fase.

- **Interferensi:**  
  - Ketika dua sinyal dengan frekuensi yang sama tetapi pergeseran fase berbeda dijumlahkan, hasilnya dapat berupa interferensi konstruktif (penguatan) atau destruktif (pelemahan), tergantung pada perbedaan fase antara kedua sinyal tersebut.

## 4. Tugas Lanjutan: Kombinasi Sinyal
Kita akan membuat sinyal baru \\( y_5 \\) yang merupakan penjumlahan dari y1 dan y2:
\\[
y_5 = y_1 + y_2
\\]


```python
# Membuat sinyal y5
y5 = y1 + y2

# Plot sinyal y5
plt.figure(figsize=(10, 4))
plt.plot(t, y5, color='purple')
plt.title('Sinyal y5 = y1 + y2')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```

### Diskusi Sinyal y5
Ketika dua sinyal dijumlahkan:
- Jika keduanya sedang di puncak pada waktu yang sama, puncak akan lebih tinggi (interferensi konstruktif).  
- Jika satu sinyal berada di puncak dan yang lain di lembah, mereka akan saling melemahkan (interferensi destruktif).  
- Karena y1 dan y2 memiliki frekuensi yang berbeda (3 Hz vs 4 Hz), bentuk gelombang y5 akan berubah-ubah sepanjang waktu, menimbulkan pola yang kompleks.
